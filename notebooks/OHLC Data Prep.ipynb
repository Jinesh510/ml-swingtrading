```python
import os
import shutil
import pandas as pd
```

```python
from pathlib import Path
```

```python
source_folder = '/Users/jineshshah/Documents/eod_data_may_2025/daily/'         # replace this with actual path
```

```python
data_folder_path = Path('/Users/jineshshah/Documents/algo_trading_project/data/')
```

```python
data_folder_path
```

```python
nse_raw_folder = data_folder_path/'nse_raw'
```

```python
os.makedirs(nse_raw_folder, exist_ok=True)
```

```python
tickers_df = pd.read_csv(data_folder_path/'nse_250_updated.csv')  # replace with actual filename if needed
```

```python
tickers_df
```

```python
ticker_list = tickers_df['Ticker'].to_list()
```

```python
ticker_list
```

```python
for ticker in ticker_list:
    file_name = f"{ticker}.csv"
    source_file = os.path.join(source_folder, file_name)
    destination_file = os.path.join(nse_raw_folder, file_name)

    if os.path.exists(source_file):
        shutil.copy2(source_file, destination_file)
        print(f"✅ Copied: {file_name}")
    else:
        print(f"❌ Missing: {file_name}")

print("✅ Done copying NSE 250 files.")
```
