```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bucket_name = "IT_services"
```

```python
folder_name = "../outputs/"+bucket_name+"/feature_importance/"
```

```python
file_name = folder_name + bucket_name + str("_feature_importance_regression.png")
```

```python
file_name
```

```python
classifier_title = str("Feature Importance - ") + bucket_name + str(" Regression Model")
```

```python
model_path = "models/" + bucket_name
```

```python
model_path
```

```python
# üîÅ Update this if your model path is different
classifier_model = joblib.load(f"../{model_path}/model_regression.pkl")
```

```python
feature_importances = pd.Series(classifier_model.feature_importances_, index=classifier_model.booster_.feature_name())
```

```python
feature_importances = feature_importances.sort_values(ascending=True)
```

```python
plt.figure(figsize=(8, 10))
feature_importances.plot(kind='barh')
plt.title(classifier_title)
plt.xlabel("Importance")
plt.tight_layout()
plt.grid(True)
plt.savefig(file_name, dpi=300)  # Save the plot as PNG
plt.show()
```
