```python
import yfinance as yf
import pandas as pd
```

```python
from pathlib import Path
```

```python
file_path = Path("data") / "nse_250_stocks_list.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def classify_market_cap(market_cap_cr):
    if market_cap_cr < 10000:
        return "Small Cap"
    elif 10000 <= market_cap_cr < 30000:
        return "Mid Cap"
    else:
        return "Large Cap"
```

```python
symbols = df["Ticker_Yahoo"].tolist()
```

```python
results = []
```

```python
for symbol in symbols[:5]:
    try:
        info = yf.Ticker(symbol).info
        mc_cr = info["marketCap"] / 1e7 if "marketCap" in info else None
        results.append({
            "Ticker_Yahoo": symbol,
            "Market_Cap_Cr": round(mc_cr, 2) if mc_cr else None,
            "Market_Cap_Bucket": classify_market_cap(mc_cr) if mc_cr else None,
            "Yahoo_Sector": info.get("sector"),
            "PE_Ratio": info.get("trailingPE"),
            "PB_Ratio": info.get("priceToBook"),
            "ROE": info.get("returnOnEquity"),
            "Debt_to_Equity": info.get("debtToEquity"),
            "EPS": info.get("trailingEps")
        })
    except Exception as e:
        print(f"Error fetching {symbol}: {e}")
```
