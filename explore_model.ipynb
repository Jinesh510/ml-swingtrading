```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scrip_name = "MIDCAP_IT"
```

```python
scrip_name = "model_classifier"
```

```python
scrip_name = "model_regression"
```

```python
folder_name = "models/feature_importance"
```

```python
file_name = folder_name + str("/")+ scrip_name + str("_feature_importance.png")
```

```python
file_name
```

```python
title = str("Feature Importance - ") + scrip_name + str(" Model")
```

```python
# üîÅ Update this if your model path is different
model = joblib.load(f"models/{scrip_name}.pkl")
```

```python
feature_importances = pd.Series(model.feature_importances_, index=model.booster_.feature_name())
```

```python
feature_importances = feature_importances.sort_values(ascending=True)
```

```python
plt.figure(figsize=(8, 10))
feature_importances.plot(kind='barh')
plt.title(title)
plt.xlabel("Importance")
plt.tight_layout()
plt.grid(True)
plt.savefig(file_name, dpi=300)  # Save the plot as PNG
plt.show()
```
